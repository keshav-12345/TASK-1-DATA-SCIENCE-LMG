# Task 1 : Image to Pencil Sketch with Python

```python
# Imprting the Required Libraries
from PIL import Image
from IPython.display import display
import cv2
```

```python
image = cv2.imread('D:/Internship/Letsgrowmore/Data Science Intern/Tasks/Task 1/4. Image to Pencil Sketch with Python/Girl.jpg', 1)
image = cv2.resize(image, (340, 250))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

# Converting Original Image into Grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

# Transformation of the greyscale image into negative using bitwise_not operator

```python
image_inv = cv2.bitwise_not(gray_image)
display(Image.fromarray(image_inv))
```

```python
image_smooth = cv2.GaussianBlur(image_inv, (21, 21), 0) # To smoothe the imahe using Gaussian function
display(Image.fromarray(image_smooth))
```

```python
def dodge(gray_image, image_smooth):
  image_smooth_inv = 255 - image_smooth
  return cv2.divide(gray_image, image_smooth_inv, scale=256.0)
```

# Pencil Sketch from Original Image

```python
image_final = dodge(gray_image, image_smooth)
display(Image.fromarray(image_final))
```

# Original Vs Sketch Image

```python
display(Image.fromarray(image))
display(Image.fromarray(image_final))
```
